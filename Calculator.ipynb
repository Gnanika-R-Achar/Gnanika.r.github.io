```python
# Method 1: Using modulus operator
num = int(input("Enter a number: "))

if num % 2 == 0:
    print("Even number")
else:
    print("Not even")
```

```python
# Method 2: Using bitwise operator
num = int(input("Enter a number: "))

if (num & 1) == 0:
    print("Even number")
else:
    print("Not even")
```

```python
# Method 1: Using modulus
num = int(input("Enter a number: "))

if num % 2 != 0:
    print("Odd number")
else:
    print("Not odd")
```

```python
# Method 2: Using bitwise operator
num = int(input("Enter a number: "))

if (num & 1) == 1:
    print("Odd number")
else:
    print("Not odd")
```

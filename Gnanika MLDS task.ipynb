```python
# Ask the user for a password
password = input("Enter a password: ")

# Check conditions
length_ok = len(password) >= 8
has_upper = any(char.isupper() for char in password)
has_lower = any(char.islower() for char in password)
has_digit = any(char.isdigit() for char in password)

# Validate password
if length_ok and has_upper and has_lower and has_digit:
    print("Strong password!")
else:
    print("Weak password")
    
    if not length_ok:
        print("- Must be at least 8 characters long")
    if not has_upper:
        print("- Must contain at least one uppercase letter")
    if not has_lower:
        print("- Must contain at least one lowercase letter")
    if not has_digit:
        print("- Must contain at least one digit")
```

```python
# Ask the user for a string
text = input("Enter a string: ")

# Create empty dictionary
frequency = {}

# Loop through each character
for char in text:
    if char in frequency:
        frequency[char] += 1
    else:
        frequency[char] = 1

# Print the result
for char in frequency:
    print(char + ":", frequency[char])
```

```python
# Ask the user for a paragraph
text = input("Enter a paragraph: ")

# Convert entire text to lowercase first (for clean formatting)
text = text.lower()

# Initialize result and a flag
result = ""
capitalize_next = True

# Loop through each character
for char in text:
    if capitalize_next and char.isalpha():
        result += char.upper()
        capitalize_next = False
    else:
        result += char

    # If sentence ends, capitalize next letter
    if char in ".!?":
        capitalize_next = True

# Print the result
print("Formatted Paragraph:")
print(result)
```

```python
def mask_email(email):
    if "@" not in email:
        return email
    
    username, domain = email.split("@")
    if len(username) <= 2:
        return email
    
    # Keep first and last letter of username, mask middle
    masked_username = username[0] + "*" * (len(username)-2) + username[-1]
    return masked_username + "@" + domain

def mask_phone(phone):
    # Remove any non-digit characters first
    digits = ''.join(filter(str.isdigit, phone))
    if len(digits) < 4:
        return phone
    
    # Keep last 4 digits, mask the rest
    return "*" * (len(digits)-4) + digits[-4:]

# Test
email = input("Enter email: ")
phone = input("Enter phone number: ")

print(f"Masked email: {mask_email(email)}")
print(f"Masked phone: {mask_phone(phone)}")
```

```python
# Ask the user for text
text = input("Enter a paragraph: ")

# Total characters (including spaces)
total_chars_with_spaces = len(text)

# Total characters (excluding spaces)
total_chars_without_spaces = len(text.replace(" ", ""))

# Total words
words = text.split()
total_words = len(words)

# Average word length
if total_words > 0:
    avg_word_length = sum(len(word) for word in words) / total_words
else:
    avg_word_length = 0

# Most common character (excluding spaces)
frequency = {}
for char in text:
    if char != " ":
        frequency[char] = frequency.get(char, 0) + 1

most_common_char = max(frequency, key=frequency.get) if frequency else None

# Number of sentences (based on ., !, ?)
sentence_count = sum(text.count(p) for p in ".!?")

# Print results
print("\n--- Text Analysis ---")
print("Total characters (with spaces):", total_chars_with_spaces)
print("Total characters (without spaces):", total_chars_without_spaces)
print("Total words:", total_words)
print("Average word length:", round(avg_word_length, 2))
print("Most common character:", most_common_char)
print("Number of sentences:", sentence_count)
```

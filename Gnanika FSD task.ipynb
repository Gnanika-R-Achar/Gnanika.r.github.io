```python
# Step 1: Define the function
def check_even_odd(num):
    if num % 2 == 0:
        return "Even"
    else:
        return "Odd"

# Step 2: Take input from the user
number = int(input("Enter a number: "))

# Step 3: Call the function
result = check_even_odd(number)

# Step 4: Print the result
print("The number is:", result)
```

```python
# Step 1: Define the function
def sum_of_n_numbers(n):
    total = 0
    for i in range(1, n + 1):
        total += i
    return total

# Step 2: Take input from the user
number = int(input("Enter a number: "))

# Step 3: Call the function
result = sum_of_n_numbers(number)

# Step 4: Print the result
print("The sum of first", number, "numbers is:", result)
```

```python
# Step 1: Create a list of 5 numbers
numbers = [10, 20, 30, 40, 50]

# Step 2: Add a new number to the list
numbers.append(60)

# Step 3: Remove one number from the list
numbers.remove(20)

# Step 4: Print the updated list
print("Updated List:", numbers)

# Step 5: Print the maximum and minimum values
print("Maximum value:", max(numbers))
print("Minimum value:", min(numbers))
```

```python
# Step 1: Create a dictionary with student names and their marks
student_marks = {
    "Asha": 85,
    "Ravi": 92,
    "Kiran": 78,
    "Meena": 88
}

# Step 2: Find the student who scored the highest marks
top_student = max(student_marks, key=student_marks.get)

# Step 3: Print the name of the top student
print("Student with highest marks:", top_student)

# Step 4: Add a new student with marks
student_marks["Arjun"] = 95

# Step 5: Print the updated dictionary
print("Updated Dictionary:", student_marks)
```

```python
# Step 1: Create a list containing duplicate values
numbers = [10, 20, 30, 20, 40, 10, 50, 30]

# Step 2: Convert the list into a set
unique_numbers = set(numbers)

# Step 3: Print the unique values
print("Unique values:", unique_numbers)
```

```python
# Step 1: Create a tuple with 5 numbers
numbers = (10, 20, 30, 40, 50)

# Step 2: Convert tuple to list (since tuples are immutable)
numbers_list = list(numbers)

# Step 3: Change one value
numbers_list[2] = 99   # Changing the third element

# Step 4: Convert list back to tuple
numbers = tuple(numbers_list)

# Step 5: Print the updated tuple
print("Updated Tuple:", numbers)
```

```python
# Step 1: Create a list with 3 elements
numbers = [10, 20, 30]

# Step 2: Use try block to access the 5th element
try:
    print("Accessing 5th element:", numbers[4])  # Index 4 means 5th element

# Step 3: Handle the error using except
except IndexError:
    print("Error: The element you are trying to access does not exist in the list.")
```

```python
# Step 1: Create a list of 10 numbers
numbers = [12, 25, 18, 40, 33, 27, 50, 16, 29, 21]

# Step 2: Function to calculate average
def calculate_average(nums):
    total = sum(nums)
    return total / len(nums)

# Step 3: Function to find maximum value
def find_maximum(nums):
    return max(nums)

# Step 4: Function to find minimum value
def find_minimum(nums):
    return min(nums)

# Step 5: Call the functions
average = calculate_average(numbers)
maximum = find_maximum(numbers)
minimum = find_minimum(numbers)

# Step 6: Print the results clearly
print("Numbers:", numbers)
print("Average:", average)
print("Maximum value:", maximum)
print("Minimum value:", minimum)
```

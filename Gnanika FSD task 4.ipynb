```python
# Problem 1:

class BankAccount:
    def __init__(self, name, branch, balance=0):
        self.name = name
        self.branch = branch
        self.balance = balance

    # Deposit money
    def credit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"₹{amount} credited successfully.")
        else:
            print("Invalid amount! Cannot credit zero or negative value.")

    # Withdraw money
    def debit(self, amount):
        if amount > 0:
            if amount <= self.balance:
                self.balance -= amount
                print(f"₹{amount} debited successfully.")
            else:
                print("Insufficient balance!")
        else:
            print("Invalid amount! Cannot debit zero or negative value.")

    # Get current balance
    def get_balance(self):
        print(f"Current balance: ₹{self.balance}")
        return self.balance

    # Get account details
    def account_details(self):
        print("----- Account Details -----")
        print(f"Name   : {self.name}")
        print(f"Branch : {self.branch}")
        print(f"Balance: ₹{self.balance}")
        print("----------------------------")

# Example Usage
account1 = BankAccount("Gnanika Achar", "MG Road Branch", 5000)

# Credit and Debit transactions
account1.credit(2000)
account1.debit(1500)
account1.debit(7000)  # Should show insufficient balance
account1.get_balance()
account1.account_details()
```

```python
# Problem 2:

def nums(*args):
    total = 0
    for i in args:
        total += i
    return total

n = int(input("Enter repetitions needed: "))
values = []

count = 0
while count < n:
    value = int(input(f"Enter number {count + 1}: "))
    values.append(value)
    count += 1

print("Sum of numbers:", nums(*values))
```

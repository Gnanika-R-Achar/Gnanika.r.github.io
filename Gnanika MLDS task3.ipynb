```python
def shift_left():
    # Create the list
    items = [10, 20, 30, 40, 50]
    
    # Store first element
    first = items[0]
    
    # Shift elements left
    for i in range(len(items) - 1):
        items[i] = items[i + 1]
    
    # Move first element to end
    items[-1] = first
    
    # Print result
    print(items)

# Call the function
shift_left()
```

```python
def shift_right():
    # Create the list
    items = [10, 20, 30, 40, 50]
    
    # Store last element
    last = items[-1]
    
    # Shift elements right
    for i in range(len(items) - 1, 0, -1):
        items[i] = items[i - 1]
    
    # Move last element to beginning
    items[0] = last
    
    # Print result
    print(items)

# Call the function
shift_right()
```

```python
def check_palindrome():
    # Create the list
    word = ['r', 'a', 'c', 'e', 'c', 'a', 'r']
    
    is_palindrome = True
    
    # Check from both ends
    for i in range(len(word) // 2):
        if word[i] != word[len(word) - 1 - i]:
            is_palindrome = False
            break
    
    # Print result
    if is_palindrome:
        print("Palindrome")
    else:
        print("Not palindrome")

# Call the function
check_palindrome()
```

```python
def merge_sorted_lists():
    # Create the sorted lists
    list_a = [1, 3, 5, 7, 9]
    list_b = [2, 4, 6, 8, 10]
    
    merged = []
    i = 0
    j = 0
    
    # Merge both lists
    while i < len(list_a) and j < len(list_b):
        if list_a[i] < list_b[j]:
            merged.append(list_a[i])
            i += 1
        else:
            merged.append(list_b[j])
            j += 1
    
    # Add remaining elements
    while i < len(list_a):
        merged.append(list_a[i])
        i += 1
        
    while j < len(list_b):
        merged.append(list_b[j])
        j += 1
    
    # Print result
    print(merged)

# Call the function
merge_sorted_lists()
```

```python
def student_grade_calculator():
    students = [
        ["Alice", 85, 90, 78, 92],
        ["Bob", 88, 75, 82, 79],
        ["Charlie", 92, 95, 88, 94]
    ]
    
    for student in students:
        name = student[0]
        grades = student[1:]
        
        # Calculate average
        average = sum(grades) / len(grades)
        
        # Assign letter grade
        if average >= 90:
            grade = "A"
        elif average >= 80:
            grade = "B"
        elif average >= 70:
            grade = "C"
        else:
            grade = "F"
        
        # Print result
        print(f"{name} - Average: {average:.2f} - Grade: {grade}")

# Call the function
student_grade_calculator()
```

```python
def rotate_playlist():
    # Create the list
    playlist = ["song1", "song2", "song3", "song4", "song5"]
    
    # Number of rotations
    rotations = 2
    
    # Rotate first two elements to the end
    for _ in range(rotations):
        first = playlist[0]
        
        # Shift left
        for i in range(len(playlist) - 1):
            playlist[i] = playlist[i + 1]
        
        playlist[-1] = first
    
    # Print result
    print(playlist)

# Call the function
rotate_playlist()
```

```python
def tic_tac_toe():
    board = [
        ["X", "O", "X"],
        ["O", "X", "O"],
        ["X", "", "O"]
    ]
    
    # Print board nicely
    print("Board:")
    for row in board:
        print(" | ".join(cell if cell != "" else " " for cell in row))
        print("-" * 9)
    
    # Count X's and O's
    x_count = 0
    o_count = 0
    
    for row in board:
        for cell in row:
            if cell == "X":
                x_count += 1
            elif cell == "O":
                o_count += 1
    
    print("X count:", x_count)
    print("O count:", o_count)
    
    # Check winner
    winner = None
    
    # Check rows
    for row in board:
        if row[0] == row[1] == row[2] != "":
            winner = row[0]
    
    # Check columns
    for col in range(3):
        if board[0][col] == board[1][col] == board[2][col] != "":
            winner = board[0][col]
    
    # Check diagonals
    if board[0][0] == board[1][1] == board[2][2] != "":
        winner = board[0][0]
        
    if board[0][2] == board[1][1] == board[2][0] != "":
        winner = board[0][2]
    
    if winner:
        print("Winner:", winner)
    else:
        print("No winner")

# Call function
tic_tac_toe()
```

```python
def show_tasks(tasks):
    if not tasks:
        print("No tasks available.")
    else:
        print("\nPending Tasks:")
        for i in range(len(tasks)):
            print(f"{i + 1}. {tasks[i]}")


def show_completed(completed):
    if not completed:
        print("No completed tasks.")
    else:
        print("\nCompleted Tasks:")
        for i in range(len(completed)):
            print(f"{i + 1}. {completed[i]}")


def add_task(tasks):
    task = input("Enter task: ")
    position = input("Add to (1) Beginning or (2) End? ")
    
    if position == "1":
        tasks.insert(0, task)
    elif position == "2":
        tasks.append(task)
    else:
        print("Invalid choice. Adding to end.")
        tasks.append(task)
    
    print("Task added successfully.")


def remove_task(tasks):
    if not tasks:
        print("No tasks to remove.")
        return
    
    choice = input("Remove by (1) Name or (2) Position? ")
    
    if choice == "1":
        name = input("Enter task name: ")
        if name in tasks:
            tasks.remove(name)
            print("Task removed.")
        else:
            print("Task not found.")
    
    elif choice == "2":
        pos = int(input("Enter task number: "))
        if 1 <= pos <= len(tasks):
            removed = tasks.pop(pos - 1)
            print(f"Removed: {removed}")
        else:
            print("Invalid position.")
    else:
        print("Invalid choice.")


def mark_complete(tasks, completed):
    if not tasks:
        print("No tasks to mark complete.")
        return
    
    pos = int(input("Enter task number to mark complete: "))
    
    if 1 <= pos <= len(tasks):
        task = tasks.pop(pos - 1)
        completed.append(task)
        print("Task marked as complete.")
    else:
        print("Invalid position.")


def todo_manager():
    tasks = []
    completed = []
    
    while True:
        print("\n==== TO-DO LIST MENU ====")
        print("1. Add Task")
        print("2. Remove Task")
        print("3. Show Tasks")
        print("4. Mark Task as Complete")
        print("5. Show Completed Tasks")
        print("6. Exit")
        
        choice = input("Enter choice: ")
        
        if choice == "1":
            add_task(tasks)
        elif choice == "2":
            remove_task(tasks)
        elif choice == "3":
            show_tasks(tasks)
        elif choice == "4":
            mark_complete(tasks, completed)
        elif choice == "5":
            show_completed(completed)
        elif choice == "6":
            print("Exiting program...")
            break
        else:
            print("Invalid choice. Try again.")


# Run program
todo_manager()
```

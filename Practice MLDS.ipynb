```python
# Check if triangle is equilateral, isosceles, or scalene
s1 = float(input("Enter side 1: "))
s2 = float(input("Enter side 2: "))
s3 = float(input("Enter side 3: "))

# Check if triangle is valid
if s1 + s2 > s3 and s2 + s3 > s1 and s1 + s3 > s2:
    if s1 == s2 == s3:
        print("Equilateral Triangle")
    elif s1 == s2 or s2 == s3 or s1 == s3:
        print("Isosceles Triangle")
    else:
        print("Scalene Triangle")
else:
    print("Invalid triangle")
```

```python
# Check if a number is prime
num = int(input("Enter a number: "))
is_prime = True

if num < 2:
    is_prime = False
else:
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            is_prime = False
            break

if is_prime:
    print(f"{num} is a prime number")
else:
    print(f"{num} is not a prime number")
```

```python
# Check if number is Armstrong (153 = 1³ + 5³ + 3³)
num = int(input("Enter a number: "))
original = num
num_digits = len(str(num))
sum_of_powers = 0

while num > 0:
    digit = num % 10
    sum_of_powers += digit ** num_digits
    num //= 10

if sum_of_powers == original:
    print(f"{original} is an Armstrong number")
else:
    print(f"{original} is not an Armstrong number")
```

```python
# Rewrite the calculator from Problem 10 using match-case
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
op = input("Enter operator (+, -, *, /): ")

match op:
    case '+':
        print(f"{num1} + {num2} = {num1 + num2}")
    case '-':
        print(f"{num1} - {num2} = {num1 - num2}")
    case '*':
        print(f"{num1} * {num2} = {num1 * num2}")
    case '/':
        if num2 != 0:
            print(f"{num1} / {num2} = {num1 / num2}")
        else:
            print("Error: Division by zero!")
    case _:
        print("Invalid operator")
```

```python
# Print message based on HTTP status code
code = int(input("Enter HTTP status code: "))

match code:
    case 200:
        print("OK - Request successful")
    case 201:
        print("Created - Resource created successfully")
    case 301:
        print("Moved Permanently - Resource has moved")
    case 302:
        print("Found - Resource found temporarily")
    case 400:
        print("Bad Request - Server couldn't understand request")
    case 401:
        print("Unauthorized - Authentication required")
    case 403:
        print("Forbidden - Access denied")
    case 404:
        print("Not Found - Resource doesn't exist")
    case 500:
        print("Internal Server Error - Server error")
    case 502:
        print("Bad Gateway - Invalid response from upstream")
    case 503:
        print("Service Unavailable - Server is down")
    case _:
        print(f"{code} - Unknown status code")
```

```python
# Calculate area based on shape input
shape = input("Enter shape (circle, square, rectangle): ").lower()

match shape:
    case "circle":
        radius = float(input("Enter radius: "))
        area = 3.14159 * radius ** 2
        print(f"Area of circle: {area:.2f}")
    
    case "square":
        side = float(input("Enter side length: "))
        area = side ** 2
        print(f"Area of square: {area}")
    
    case "rectangle":
        length = float(input("Enter length: "))
        width = float(input("Enter width: "))
        area = length * width
        print(f"Area of rectangle: {area}")
    
    case _:
        print("Invalid shape")
```

```python
# Check if a string is palindrome (reads same forwards and backwards)
text = input("Enter a string: ").lower()
is_palindrome = True

# Remove spaces and punctuation (optional)
clean_text = ''.join(char for char in text if char.isalnum())

for i in range(len(clean_text) // 2):
    if clean_text[i] != clean_text[len(clean_text) - 1 - i]:
        is_palindrome = False
        break

if is_palindrome:
    print(f"'{text}' is a palindrome")
else:
    print(f"'{text}' is not a palindrome")
```

```python
# Password must be at least 8 characters, contain a number and a special char
password = input("Enter password: ")
has_number = False
has_special = False
special_chars = "!@#$%^&*()_+-=[]{}|;:,.<>?"

if len(password) < 8:
    print("Password must be at least 8 characters long")
else:
    for char in password:
        if char.isdigit():
            has_number = True
        if char in special_chars:
            has_special = True
    
    if has_number and has_special:
        print("Password is valid!")
    else:
        if not has_number:
            print("Password must contain at least one number")
        if not has_special:
            print("Password must contain at least one special character")
```

```python
# Same as Problem 23 but give hints: "too high", "too low", and count attempts
import random
secret = random.randint(1, 50)
attempts = 0
max_attempts = 7

print("I'm thinking of a number between 1 and 50")
print(f"You have {max_attempts} attempts")

while attempts < max_attempts:
    guess = int(input("Your guess: "))
    attempts += 1
    
    if guess == secret:
        print(f"Correct! You guessed it in {attempts} attempts!")
        break
    elif guess < secret:
        print(f"Too low! Attempts left: {max_attempts - attempts}")
    else:
        print(f"Too high! Attempts left: {max_attempts - attempts}")
else:
    print(f"Game over! The number was {secret}")
```

```python
# Implement Rock Paper Scissors game
import random

choices = ["rock", "paper", "scissors"]
computer = random.choice(choices)
user = input("Enter rock, paper, or scissors: ").lower()

if user not in choices:
    print("Invalid choice!")
else:
    print(f"Computer chose: {computer}")
    
    match (user, computer):
        # Tie cases
        case (u, c) if u == c:
            print("It's a tie!")
        
        # User win cases
        case ("rock", "scissors"):
            print("You win! Rock crushes scissors")
        case ("paper", "rock"):
            print("You win! Paper covers rock")
        case ("scissors", "paper"):
            print("You win! Scissors cut paper")
        
        # Computer win cases
        case _:
            print(f"You lose! {computer} beats {user}")
```

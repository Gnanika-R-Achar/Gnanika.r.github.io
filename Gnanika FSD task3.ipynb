```python
def register_user(reg_no, name, email, password, confirm_password):
    # Check if passwords match
    if password != confirm_password:
        print("Error: Passwords do not match!")
        return
    
    # Check if email contains '.com'
    if not email.endswith(".com"):
        print("Error: Email must end with '.com'")
        return

    # Store the details in a list
    user_data = [reg_no, name, (email,), password]
    
    # Display the registered user data
    print("Registration Successful!")
    print("User Details:")
    print(f"Register No: {user_data[0]}")
    print(f"Name: {user_data[1]}")
    print(f"Email: {user_data[2][0]}")  # tuple inside list
    print(f"Password: {user_data[3]}")

# Example usage:
register_user("101", "Gnanika R", "gnanika@gmail.com", "12345", "12345")
register_user("102", "Ravi", "ravi@gmail.com", "abc", "abc")  # will show email error
register_user("103", "Anu", "anu@gmail.com", "pass", "wrongpass")  # will show password error
```
